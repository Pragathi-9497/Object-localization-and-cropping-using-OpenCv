```python
import os
import cv2
import pylab as py
import numpy as np
from matplotlib import pyplot as plt
```

```python
image=cv2.imread(r'C:\Users\Lenovo\Documents\dataset\coffee_dataset2\test_test\beans_for_detection\6.jpg')
plt.imshow(image)
```

```python
gray_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
thresh_image=cv2.threshold(gray_image, 0,255, cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)[1]
plt.imshow(thresh_image)
```

```python
count,_=cv2.findContours(thresh_image,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
```

```python
c=max(count,key=cv2.contourArea)
x,y,h,w=cv2.boundingRect(c)
x=x-200
y=y-200
w=w+250
h=w+250
cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,0),10)
cropped_image=image[y:y+h,x:x+w]
plt.imshow(cropped_image)
```
